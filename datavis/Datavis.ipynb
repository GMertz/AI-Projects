```python
import numpy as np

import pandas as pd
data = pd.read_csv('USvideos.csv', delimiter=',')


#data.head()
#data.info()
import datetime as dt
std = np.linalg.norm([(a/(b+1)) for a,b in zip(data['likes'],data['dislikes'])])
print(std)
def formatTime(point):
    return dt.datetime.strptime(point, "%y.%d.%m")
    #return dt.datetime.strptime(point, "%Y-%m-%dT%H:%M:%S.000Z")

    
#print(formatTime(data['publish_time'][0]).time())
#data.describe()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib as mpl
import numpy as np
import scipy


x = [formatTime(point) for point in data['trending_date']]
y = [views/1000 for views in data['views']]
color = [a/(a+b+1) for a,b in zip(data['likes'],data['dislikes'])]

size = data['comment_count']/100
fig = plt.scatter(x,y,s=size,c=color,cmap=cm.summer_r,alpha=.5)


cb = plt.colorbar()
cb.set_label("Likes / Dislikes")

firstdate = formatTime(data['trending_date'][0])
lastdate = formatTime(data['trending_date'][20965])

plt.rcParams["figure.figsize"] = (20,20)
plt.xlim(firstdate,lastdate)
plt.xticks(rotation=50)
plt.xlabel("Trending Date")
plt.ylabel("Views \n(in thousands)")
plt.suptitle('Trending Youtube Videos from November to February')


plt.show()
```
