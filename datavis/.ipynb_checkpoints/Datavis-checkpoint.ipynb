```python
import numpy as np

import pandas as pd
data = pd.read_csv('USvideos.csv', delimiter=',')


#data.head()
#data.info()
import datetime as dt

def formatTime(point):
    return dt.datetime.strptime(point, "%y.%m.%d")
    #return dt.datetime.strptime(point, "%Y-%m-%dT%H:%M:%S.000Z")

    
#print(formatTime(data['publish_time'][0]).time())
#data.describe()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib as mpl
import numpy as np


x = [formatTime(point) for point in data['publish_time']]
y = [views/1000 for views in data['views']]
color = [a/(b+1) for a,b in zip(data['likes'],data['dislikes'])]

size = data['comment_count']/100
norm = mpl.colors.Normalize(vmin=.4,vmax=2.)
fig = plt.scatter(x,y,s=size,c=color,cmap=cm.RdYlGn,alpha=.6,norm=norm)


cb = plt.colorbar()
cb.set_label("Likes / Dislikes")

firstdate = formatTime(data['publish_time'][0])
lastdate = formatTime(data['publish_time'][20965])

plt.rcParams["figure.figsize"] = (20,20)
plt.xlim(firstdate,lastdate)
plt.xticks(rotation=50)
plt.xlabel("Publish Time")
plt.ylabel("Views \n(in thousands)")
plt.suptitle('Trending Youtube Videos from November to February')


plt.show()
```
