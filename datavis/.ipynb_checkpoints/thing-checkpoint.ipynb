```python
import numpy as np

import pandas as pd
data = pd.read_csv('USvideos.csv', delimiter=',')


#data.head()
data.info()
import datetime as dt
def formatTime(point):
    #point = data['publish_time'][0]
    #print(point)
    #print(point[point.index('T')+1:point.index('Z')])
    return dt.datetime.strptime(\
        point[point.index('T')+1:point.index('Z')-4],"%H:%M:%S")
    #print(dt.datetime.strptime(point,"%H:%M:%S").time())

    
#print(formatTime(data['publish_time'][0]).time())
#data.describe()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
x = [formatTime(point).time() \
     for point in data['publish_time']]
y = data['views']
color = [a/(b+1) for a,b in zip(data['likes'],data['dislikes'])]
fig = plt.scatter(x,y,s=10,c=color,cmap=cm.seismic,alpha=.1)
plt.colorbar()
plt.xlabel("Publish Time")
plt.ylabel("Views")
plt.ylim([70000,2000000])




plt.show()
```
